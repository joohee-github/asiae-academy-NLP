```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import tensorflow as tf


seed = 0
np.random.seed(seed)
# random seed 설정  tf 2.0버전 code
tf.random.set_seed(seed)


# 환자 데이터 불러오기
Data_set =  np.loadtxt('C:/Users/user/study1/머신러닝/data/ThoraricSurgery.csv',delimiter = ',')


# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장한다.
X = Data_set[:,:17]
Y = Data_set[:,[17]]

# 케라스에서 층 구성
model = Sequential()


# model.add()  코드를 통해 층을 단계적으로 추가
# Dense()
# 첫번째 인자 = 출력 뉴런의 수
# 두 번째 인자 input_dim = 입력 뉴런의 수
# activation = 활성화 함수
model.add(Dense(30, input_dim = 17, activation = 'sigmoid'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'mean_squared_error',optimizer = 'adam',
              metrics = ['accuracy'])
model.fit(X,Y,epochs = 30, batch_size = 10)

print('\n Accuracy: %.4f' % (model.evaluate(X,Y)[1]))

```

- model.complie() :  값을 파라미터로 넣어주는 것이아니라 여러가지 파라미터를 세팅해주는 역할
- 위에서 정해진 모델을 컴퓨터가 알아들을 수 있게 컴파일 하는 부분

- nodel.fit()  모델을 실제로 수행하는 부분
- 마지막으로 입력 데이터의 넘파이 배열을 모델의 fit메서드에 전달함으로써 학습과정이 이루어짐
- 케라스에서는 모델을 sequential 함수를 이용해서 층을 쌓듯이 사용한다.
- model.add = 새로운 층을 만들어줌.
- Dense(앞의 Node, input_dim, activation)
- 각각의 층은 Dense라는 함수를 통해 구체적으로 구조가 결정된다.


- model.add(Dense(몇개의 노드를 만들것인지를 결정, 들어오는 node의 값, activation))
-  출력층 :  model.add(Dense(1, activation= 'sigmoid')
- 은닉층이 각 노드는 입력값들로부터 임의의 가중치를 가지고 각 노드로 전송되어 활성화 함수를 만난다.

- evaluate() 테스트 데이터를 통해 학습한 모델에 대한 정확도를 평가한다.
- 첫번째 데이터 = 테스트 데이터
- 두번째 인자 = 

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense

training_data = np.array([[0,0],[0,1],[1,0],[1,1]], 'float32')
target_data = np.array([[0],[1],[1],[0]],'float32')

model = Sequential()
model.add(Dense(32, input_dim = 2, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy',optimizer = 'adam',
metrics = ['binary_accuracy'])

model.fit(training_data, target_data, epochs = 1000, verbose = 2)

print(model.predict(training_data))
print('\n Accuracy: %.4f' % (model.evaluate(training_data, target_data)[1]))
```
