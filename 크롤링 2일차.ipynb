```python
def get_day_sise(code, page_num):
    base_url = 'https://finance.naver.com/item/sise_day.nhn'

    result_list = []
    for page in range(1, page_num+1):
        params = {
            'code' : code,
            'page' : page
        }

        resp = requests.get(base_url, params=params) #요청
        resp

        soup = BeautifulSoup(resp.text)   #파싱

        tr_tags = soup.select('table.type2 tr') 

        th_tags = tr_tags[0].find_all('th')

        title_list = []
        for th in th_tags:
            title_list.append(th.text)

        
        
        for tr_tag in  tr_tags:
            data_list = []
            td_tags = tr_tag.find_all('td')
            if len(td_tags)<7:
                continue   # 7개 이하면 버린다.
            
            for td in td_tags:
                data_list.append(td.text.strip())

            result_list.append(
            dict(zip(title_list, data_list))
            )
            
    return result_list

    
    
```

```python
sise = get_day_sise('005930',5)
```

```python
base_url = 'https://news.naver.com/main/list.nhn'

params = {
    'mode':'LS2D',
    'sid2':'731',
    'sid1':'105',
    'mid':'shm',
    'date':'20200731',
    'page':'1'
    
}

```

```python
resp = requests.get(base_url, params=params)
resp

soup = BeautifulSoup(resp.text)
print(soup)
```

```python
div_tags = soup.find('div',class_ = 'newsflash_body')
li_tags = div_tags.find_all('li')
len(li_tags)
```

```python
result_list = []


for li in li_tags:    
    title_tag = li.find('a')    
    title = title_tag.text
    url = title_tag.get('href')

    detail_resp = requests.get(url)

    content = detail_resp.text
    result_list.append({
        'title':title,
        'content':content
    })

```
