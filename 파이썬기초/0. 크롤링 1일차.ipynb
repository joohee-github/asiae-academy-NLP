## 블로그 제목, url, 내용 5페이지까지 함수 / 답

```python
url = "https://search.naver.com/search.naver"


page_num = int(input('Page Number?'))
blog_list = []   # 여러페이지의 값을 담기 위해 반복문 밖으로 빼줌
for i in range(page_num):

    params = {
        'where' : 'post',
        'query' : '방탄소년단',
        'start' : i *10 + 1
    }

    resp = requests.get(url, params=params)
    resp

    soup = BeautifulSoup(resp.text)

    blog_list_tags = soup.find_all('li',class_ = 'sh_blog_top')

    
    
    for blog in blog_list_tags:
        a_tag = blog.find('a',class_='sh_blog_title _sp_each_url _sp_each_title')
        blog_url = a_tag.get('href')
        title = a_tag.text
        desc = blog.find_all('dd')[1].text

        blog_list.append({
            'url':blog_url,
            'title' : title,
            'desc' : desc
        })
```

# 블로그 크롤링 함수 (페이지,쿼리)

```python
def crawl_blog(query, page_num):
    url = "https://search.naver.com/search.naver"
    
    
    blog_list = []   # 여러페이지의 값을 담기 위해 반복문 밖으로 빼줌
    for i in range(page_num):

        params = {
            'where' : 'post',
            'query' : query,
            'start' : i *10 + 1
        }

        resp = requests.get(url, params=params)
        resp

        soup = BeautifulSoup(resp.text)

        blog_list_tags = soup.find_all('li',class_ = 'sh_blog_top')



        for blog in blog_list_tags:
            a_tag = blog.find('a',class_='sh_blog_title _sp_each_url _sp_each_title')
            blog_url = a_tag.get('href')
            title = a_tag.text
            desc = blog.find_all('dd')[1].text

            blog_list.append({
                'url':blog_url,
                'title' : title,
                'desc' : desc
            })

    return blog_list
```

```python
blog_search_list = crawl_blog('트와이스',3)
print(len(blog_search_list))
blog_search_list
```

# 뉴스 상세페이지/답

```python
url = "https://search.naver.com/search.naver"

params = {
    'where' : 'news',
    'query' : '방탄소년단'
}
resp = requests.get(url, params=params)
resp

soup = BeautifulSoup(resp.text)

a_tags = soup.select('ul.type01>li>dl>dt>a._sp_each_title')
len(a_tags)
```

```python
result_list = []

for a_tag in a_tags:
    title = a_tag.text
    news_url = a_tag.get('href')
    
    detail_resp = requests.get(news_url)  #######새로 요청
    
    content = detail_resp.text
    result_list.qppend({
        'title' : title,
        'url' : news_url,
        'content' : content
    })
```
