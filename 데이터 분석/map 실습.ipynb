```python
# figure에 이름 정하기

import numpy as np

# 데이터 생성
x = np.arange(-5, 5, 0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2 # NumPy에서 cos()는 np.cos()으로 입력

# figure에 이름 정하기

plt.figure(1) # 1번 그래프 창을 생성함
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그림 

plt.figure(2) # 2번 그래프 창을 생성함
plt.plot(x, y2) # 지정된 그래프 창에 그래프를 그림

plt.figure(1) # 이미 생성된 1번 그래프 창을 지정함
plt.plot(x, y2) # 지정된 그래프 창에 그래프를 그림

plt.figure(2) # 이미 생성된 2번 그래프 창을 지정함
plt.clf() # clear figure 2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그림

plt.show()

## figure은 한 번 그려지면 모두 초기화됨

```

```python
import numpy as np

x = np.arange(0,10,0.1)
y1 =  0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2  ##????

# 지수
y4 = 10*np.exp(-x) + 1

#sunplot 변수간의 컬럼간의 상관관계를 알고 싶을때

# plot을 (2,2)로 나누고 1번쨰 plot을 그린다.
plt.subplot(2,2,1)
plt.plot(x,y1)

# plot을 (2,2)로 나누고 2번쨰 plot을 그린다.
plt.subplot(2,2,2) # p는 2
plt.plot(x,y2)

plt.subplot(2,2,3) # p는 3
plt.plot(x,y3)

plt.subplot(2,2,4) # p는 4
plt.plot(x,y4)

plt.show()
```

```python
columns = ['CRIM','ZN',"INDUS",'CHAS','NOX','RM','AGE','DIS','RAD',
          'TAX',"PTRATIO",'B','LSTAT']
columns = df.columns

fig,axs = plt.subplots(7,2,figsize =(20,20))
axs = axs.ravel()
for ax,column in zip(axs,columns):
    ax.plot(df[column],y,'o')  # 산점도처럼 동그라미로 표현
    ax.set_ylabel('price',fontsize=18)
    ax.set_xlabel(column,fontsize=18) 
    
plt.tight_layout()

```

그래프 출력범위 지정

그래프 꾸미기

```python
import numpy as np
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
```

```python
plt.plot(x,y1,x,y2,x,y3,x,y4) 
#디폴트는 선그래프
```

```python
plt.plot(x, y1, 'm', x, y2,'y', x, y3, 'k', x, y4, 'c') 
#RGB 색상표를 넣어도 가능
```

```python
#선의 형태

plt.plot(x, y1, '-', x, y2, '--', x, y3, ':',  x, y4, '-.')

plt.show()
```

```python
# 마커
plt.plot(x, y1, 'o', x, y2, '^',x, y3, 's', x, y4, 'd') # square, diamond
plt.show()
```

```python
# 마커 + 라인그래프
# scatter + line (순서 상관 X)
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.show()
```

```python
import matplotlib.pyplot as plt

height = np.array([165,177,160,180, 185, 155, 172,200])
weight = np.array([62, 67,55, 74, 90, 43, 64,45])
bmi = height/weight*100

plt.scatter(height, weight,c='b',s=bmi)
plt.xlabel('Height(m)')
plt.ylabel('Weight(Kg)')
plt.title('Height & Weight')
plt.grid(True)

# 값의 크기에 따라서 점의 크기를 다르게 할 수 있음
```
