def categorizing_by_5000(data):
    if data.dtype not in [np.int64, np.float64]:
        return data
    data = data.where(lambda x : x<5000, '>=5000')
    data = data.where(lambda x : x>=5000, '<5000')    
    return data

```python
def categorizing_by_5000(data): 
    data = data.where(lambda x : x>=5000, 0)  # 5000보다 작으면 0으로 바꿈
    data = data.where(lambda x : x<5000, 1)  # 5000보다 크면 1로 바꿈
    return data
```

```python
ktx_df.iloc[:,:-2].apply(categorizing_by_5000)
```
